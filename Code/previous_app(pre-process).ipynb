```python
def balance_column(column_names, dataframe):
    for column_name in column_names:
        #print(f"\n\n--- Balancing {column_name} ---")

        # Step 1: Handle Missing Values
        # Impute missing values using the median
        dataframe[column_name].fillna(dataframe[column_name].median(), inplace=True)

        # Step 2: Handle Outliers
        # Define a function to cap outliers
        def cap_outliers(series, lower_percentile=0.01, upper_percentile=0.99):
            lower_bound = series.quantile(lower_percentile)
            upper_bound = series.quantile(upper_percentile)
            return np.clip(series, lower_bound, upper_bound)

        # Apply the function to cap outliers in the column
        dataframe[column_name] = cap_outliers(dataframe[column_name])

        # Check for Missing Values
        print(f"\nMissing Values in {column_name}:")
        print(dataframe[column_name].isnull().sum())

# List of columns to balance
columns_to_balance = [
    'AMT_ANNUITY', 'AMT_DOWN_PAYMENT', 'AMT_GOODS_PRICE', 'RATE_DOWN_PAYMENT',
    'CNT_PAYMENT', 'DAYS_FIRST_DRAWING',
    'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_LAST_DUE', 'DAYS_TERMINATION'
]

# Example usage with the filtered_previous_app dataframe
balance_column(columns_to_balance, filtered_previous_app)
```

```python
# List of columns to convert and handle outliers
columns_to_convert = [
    'DAYS_DECISION',
    'DAYS_FIRST_DRAWING',
    'DAYS_LAST_DUE_1ST_VERSION',
    'DAYS_LAST_DUE',
    'DAYS_TERMINATION'
]

# Set the threshold to 1000 (or another value you deem reasonable)
threshold = 1000

# Function to handle outliers by replacing values greater than the threshold with the median
def impute_outliers(df, column, threshold):
    # Calculate the median for the column based on values below the threshold
    median_value = df.loc[df[column] <= threshold, column].median()
    # Replace outliers (values above the threshold) with the median
    df[column] = df[column].apply(lambda x: median_value if x > threshold else x)
    return df

# Loop through each column and handle the outliers
for column in columns_to_convert:
    filtered_previous_app = impute_outliers(filtered_previous_app, column, threshold)

# Check the updated values after imputation for each column
for column in columns_to_convert:
    print(f"After imputing outliers in {column}:")
    print(filtered_previous_app[column].describe(), "\n")
```
