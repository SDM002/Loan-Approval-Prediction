```python
# Define a function to cap outliers using the IQR method
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the values outside the bounds
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

    return df

# List of columns to apply outlier capping
columns_to_cap = ['AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'DAYS_EMPLOYED', 'REGION_RATING_CLIENT']

# Apply the outlier capping function to each relevant column
for col in columns_to_cap:
    current_app = cap_outliers(current_app, col)

# Display the first few rows to ensure outliers have been capped
print(current_app[columns_to_cap].head())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline  # Import Pipeline
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE  # Import SMOTE for handling class imbalance

# Define features and target
features = [
    'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE', 'CNT_CHILDREN',
    'NAME_INCOME_TYPE', 'AMT_INCOME_TOTAL', 'OCCUPATION_TYPE', 'CNT_FAM_MEMBERS',
    'DAYS_EMPLOYED', 'REGION_RATING_CLIENT', 'REGION_RATING_CLIENT_W_CITY',
    'REG_CITY_NOT_WORK_CITY', 'LIVE_CITY_NOT_WORK_CITY', 'ORGANIZATION_TYPE'
]

target = 'TARGET'

# Split the dataset into features and target
X = current_app[features]
y = current_app[target]

# Split the dataset into training (70%) and testing + validation (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, stratify=y)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

# Apply SMOTE only to the training set to handle class imbalance
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Define preprocessor for pipeline
categorical_features = ['NAME_INCOME_TYPE', 'OCCUPATION_TYPE', 'ORGANIZATION_TYPE']
numeric_features = [feature for feature in features if feature not in categorical_features]

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# Updated models with the best hyperparameters
models = {
    "Logistic Regression": Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression(
            random_state=42,
            class_weight='balanced',
            max_iter=500,
            C=10,  # Best C value
            penalty='l2',  # Best penalty
            solver='liblinear'  # Best solver
        ))
    ]),
    "Decision Tree": Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', DecisionTreeClassifier(
            random_state=42,
            class_weight='balanced',
            max_depth=15,  # Best max_depth
            min_samples_split=2,  # Best min_samples_split
            min_samples_leaf=1  # Best min_samples_leaf
        ))
    ]),
    "Gradient Boosting": Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', GradientBoostingClassifier(
            random_state=42,
            n_estimators=100  # This can be tuned further if needed
        ))
    ])
}
```

```python
# @title Train and evaluate models using only the SMOTE-augmented training set

for name, model in models.items():
    model.fit(X_train_smote, y_train_smote)  # Train the model with the SMOTE training data
    y_train_pred = model.predict(X_train_smote)  # Predict on the training set

    print(f"\n{name} - Training Set Evaluation")
    print("Confusion Matrix :")
    print(confusion_matrix(y_train_smote, y_train_pred))

    print("Classification Report :")
    print(classification_report(y_train_smote, y_train_pred))

    print("Accuracy :", accuracy_score(y_train_smote, y_train_pred))

    # Check if the model has the predict_proba method for AUC and ROC curve
    if hasattr(model, "predict_proba"):
        y_train_pred_proba = model.predict_proba(X_train_smote)[:, 1]  # Probabilities for the positive class
        auc_score = roc_auc_score(y_train_smote, y_train_pred_proba)
        print("AUC :", auc_score)

        # Plot ROC curve
        fpr, tpr, _ = roc_curve(y_train_smote, y_train_pred_proba)
        plt.figure(figsize=(6, 6))
        plt.plot(fpr, tpr, color='orange', label=f'ROC (AUC = {auc_score:.2f})', linewidth=2)
        plt.plot([0, 1], [0, 1], linestyle='--', color='blue', label='Random Classifier')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'Receiver Operating Characteristic (ROC) Curve - {name}')
        plt.legend(loc="lower right")
        plt.show()
    else:
        print("AUC: Not applicable for this model")

```

```python
# @title Cross-validation
from sklearn.model_selection import cross_val_score, cross_val_predict, StratifiedKFold

# Define cross-validation strategy (StratifiedKFold for class balance)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation for each model
for name, model in models.items():
    scores = cross_val_score(model, X_val, y_val, cv=cv, scoring='accuracy')
    y_pred_cv = cross_val_predict(model, X_val, y_val, cv=cv)

    print(f"\n{name} - Cross-Validation:")
    print("Cross-validated Accuracy:", scores.mean(), "+/-", scores.std() * 2)

    # Evaluate classification report during cross-validation
    print("Classification Report (Cross-Validation):")
    print(classification_report(y_val, y_pred_cv))

```

```python
# @title  test the model with test data set(unseen)

for name, model in models.items():
    y_test_pred = model.predict(X_test)
    print(f"\n{name} - Test Set Evaluation:")

    # Confusion Matrix
    print("Confusion Matrix (Test Set):")
    print(confusion_matrix(y_test, y_test_pred))

    # Classification Report
    print("Classification Report (Test Set):")
    print(classification_report(y_test, y_test_pred))

    print("Accuracy (Test Set):", accuracy_score(y_test, y_test_pred))

    # AUC score on the test set (if the model supports probability prediction)
    if hasattr(model, "predict_proba"):
        y_test_pred_proba = model.predict_proba(X_test)[:, 1]
        auc_score_test = roc_auc_score(y_test, y_test_pred_proba)
        print(f"AUC (Test Set): {auc_score_test:.2f}")
    else:
        print("AUC: Not applicable for this model")

```
